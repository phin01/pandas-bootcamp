# Imports

```python
import pandas as pd
import numpy as np
```

# Load Data

```python
# Load lecture CSV file, set the proper index and remove dummy index
players = pd.read_csv('../data/soccer.csv', index_col='name')
players.head()
```

```python
players.dtypes.value_counts()
```

# Boolean Masks

What players have a market value exceeding 40m?

```python
over40m = players.market_value > 40
over40m
```

```python
players[over40m]
```

## .isin()

What players are defenders?

```python
players.position.unique()
```

```python
# Defenders are 'LB', 'CB', 'RB'
defenders = players.position.isin(['LB', 'CB', 'RB'])
defenders
```

```python
players[defenders]
```

## .between()

What players are in their 30's?

```python
thirties = players.age.between(30, 39)
thirties
```

```python
players[thirties]
```

# Multiple Conditions

Boolean masks make it easy to combine conditions

```python
# What are the defenders in their 30's?
players[defenders & thirties]
```

```python
# What players are either Spanish or German?

spanish = players.nationality == 'Spain'
german = players.nationality == 'Germany'

players[spanish | german]
```

```python
# Arsenal RBs and Chelsea GKs
arsenal = players.club == 'Arsenal'
chelsea = players.club == 'Chelsea'
rb = players.position == 'RB'
gk = players.position == 'GK'

players[(arsenal & rb) | (chelsea & gk)]
```
