# Imports

```python
import pandas as pd
import numpy as np
```

# Load Data

```python
# Load lecture CSV file, set the proper index and remove dummy index
nutrition = pd.read_csv('../data/nutrition.csv', index_col='name')
nutrition.drop('Unnamed: 0', axis=1, inplace=True)

# convert to float
nutrition = nutrition.replace(to_replace='[a-zA-Z]', value='', regex=True)
nutrition = nutrition.astype(float)

nutrition.head()
```

## Filtering in 2D

```python
nutrition.shape
```

```python
# 'like' parameter is CASE SENSITIVE
nutrition.filter(axis=0, like='Octopus')
```

```python
nutrition.filter(axis=0, like='octopus')
```

```python
# regex parameter allows for modifiers to check case insensitive values
nutrition.filter(axis=0, regex='(?i)octOPUS')
```

```python
# filter methods can be chained
(
    nutrition
    .filter(axis=0, regex='(?i)octOPUS')
    .filter(axis=1, regex='(?i)cholesterol')
)
```

```python
# you can also supply a list of items to be filtered instead of like/regex
(
    nutrition
    .filter(axis=0, regex='(?i)octOPUS')
    .filter(axis=1, items=['calories', 'sodium', 'niacin'])
)
```

## Sorting the Dataframe

```python
# Any Series can be sorted
nutrition.saturated_fat.sort_values()
```

```python
nutrition.sort_values(by=['sodium'], ascending=False).head()
```

```python
# Sort by multiple columns, and multiple orders
nutrition.sort_values(by=['cholesterol', 'sodium'], ascending=[False, True]).head()
```

## .between() method for Series

```python
nutrition.calories.head(10)
```

```python
# The .between() method will return a boolean mask
nutrition.calories.between(20, 60)
```

```python
# Same number of rows as the actual dataframe
nutrition.calories.between(20, 60).shape
```

```python
nutrition[nutrition.calories.between(20, 60)].sample(5)
```

## Min, Max, IdxMin, IdxMax

```python
# What is the highest amount of potassium from the list?
nutrition.potassium.max()
```

```python
# And what food has the max amount of potassium?
nutrition.potassium.idxmax()
```

```python
(
    nutrition
    .sort_values(by='potassium', ascending=False)
    .filter(axis=1, items=['potassium', 'sodium'])
    .head(5)
)
```

```python
# 'Good Foods' are the ones in which the Potassium/Sodium ratio is close to 16

nutrition['potassium_sodium_ratio'] = nutrition.potassium / nutrition.sodium
```

```python
(
    nutrition
    .sort_values(by='potassium', ascending=False)
    .filter(axis=1, items=['potassium', 'sodium', 'potassium_sodium_ratio'])
    .head(5)
)
```

```python
(
    nutrition[nutrition.potassium_sodium_ratio.between(15, 17)]
    .sort_values(by='potassium', ascending=False)
    .filter(axis=1, items=['potassium', 'sodium', 'potassium_sodium_ratio'])
)
```
